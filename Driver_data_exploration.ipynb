```python
driver_standings_df = pd.read_csv("../Analysis/F1_Data/driver_standings.csv")
drivers_df = pd.read_csv("../Analysis/F1_Data/drivers.csv")
result_df = pd.read_csv("../Analysis/F1_Data/results.csv")
race_df = pd.read_csv("../Analysis/F1_Data/races.csv")
```

```python
rd=pd.merge(result_df,drivers_df,on='driverId')
rdr=pd.merge(rd,race_df,on='raceId')
driver_race_df=rdr
driver_race_df
```

```python
# Create Driver column name  
driver_race_df["Driver"] = driver_race_df["forename"] + " " + driver_race_df["surname"]
driver_race_df.head()
```

```python
#move Driver name to first col
cols = driver_race_df.columns.tolist()
cols.insert(0, cols.pop(cols.index('Driver')))
driver_race_df = driver_race_df.loc[:, cols].drop('driver', axis=1)
driver_race_df.head()
```

```python
cols = []
for col in driver_race_df.columns: 
  cols.append(col) 
cols
```
