```python
import pandas as pd
import psycopg2
from psycopg2 import sql
import os
from getpass import getpass
```

```python
PGEND_POINT = 'f1stats.cos0zhtpa15v.us-east-1.rds.amazonaws.com' # End_point
PGDATABASE_NAME = 'F1_stats' # Database name
PGUSER_NAME = 'postgres' # UserName
PGPASSWORD = getpass('Enter database password')
```

```python
def connect():
    
    # Set up a connection to the postgres server.
    conn_string = "host="+ PGEND_POINT +" port="+ "5432" +" dbname="+ PGDATABASE_NAME +" user=" + PGUSER_NAME \
                  +" password="+ PGPASSWORD
    
    conn = psycopg2.connect(conn_string)
    print("Connected!")

    # Create a cursor object
    cursor = conn.cursor()
    
    return conn, cursor
```

```python
conn, cursor = connect()
```

```python
#Read in query results from each tables in the database
circuits = sql.SQL("""SELECT * FROM circuits""");
constructors = sql.SQL("""SELECT * FROM constructors""");
driver_stnd = sql.SQL("""SELECT * FROM driver_stnd""");
drivers = sql.SQL("""SELECT * FROM drivers""");
lap_times = sql.SQL("""SELECT * FROM lap_times""");
pit_stops = sql.SQL("""SELECT * FROM pit_stops""");
quali = sql.SQL("""SELECT * FROM quali""");
races = sql.SQL("""SELECT * FROM races""");
results = sql.SQL("""SELECT * FROM results""");
resultsplus = sql.SQL("""SELECT * FROM results_merged""");
sprint_rslt = sql.SQL("""SELECT * FROM sprint_rslt""");
status = sql.SQL("""SELECT * FROM status""");
final = sql.SQL("""SELECT * FROM final_merged_table""");
```

```python
#Check that query results are read in properly using circuits table
cur = conn.cursor()
cur.execute(circuits)
print(cur.fetchall())
```

```python
# Circuits dataframe

cur = conn.cursor()
cur.execute(circuits)

circuit_df = pd.DataFrame(cur.fetchall(), columns=['circuitId', 'circuitRef', 'circ_name'])
circuit_df.head(10)
# circuit_df.shape
```

```python
# Driver Standings dataframe

cur = conn.cursor()
cur.execute(driver_stnd)

driver_stnd_df = pd.DataFrame(cur.fetchall(), columns=['driverId', 'raceId', 'points', 'position', 'wins'])
driver_stnd_df.head(10)
# driver_stnd_df.shape
```

```python
# Constructors dataframe

cur = conn.cursor()
cur.execute(constructors)

constructors_df = pd.DataFrame(cur.fetchall(), columns=['constructorId','constructorRef','constname','constnationality'])
constructors_df.head(10)
# constructors_df.shape
```

```python
#Putting all 8576 rows with nulls in a dataframe for various explorative purposes
null_df= final_df[final_df.isna().sum(1) > 0]
null_df.head(10)
```

```python
#dataframe where the common null 831 rows are all dropped, but the rest left
perc = 85
limit = int(((100-perc)/100)*final_df.shape[1])
mod_final_df = final_df.dropna(axis=0, thresh=limit)
mod_final_df.isna().sum()
```
