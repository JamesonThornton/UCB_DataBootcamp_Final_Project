```python
# Import drivers_namescopy csv
drivers_names = pd.read_csv('../drivers_namescopy.csv')
df = df.merge(drivers_names, on='driverId', how='left')
df.head()
```

```python
# drop forename and surname columns
df = df.drop(['forename', 'surname'], axis=1)

# Move combined_name column to the front of the df
cols = list(df.columns)
cols = [cols[-1]] + cols[:-1]
df = df[cols]
# Change combined_name to driver_name
df = df.rename(columns={
    'combined_name': 'driver_name',
})
df.head()
```

```python
# drop decimal place for year as a whole number
df['year'] = df['year'].astype(int)
df['year'].unique()
```

```python
df['winner_name'] = df.loc[df.groupby(['year', 'races_name'])['results_position'].idxmin(), 'driver_name']

# replace NaN with N/A
df['winner_name'] = df['winner_name'].fillna('N/A')


df.head()
```
